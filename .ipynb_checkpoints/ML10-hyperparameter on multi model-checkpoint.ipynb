```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
iris=pd.read_csv("iris.csv")
df=iris.copy()
df.head()
```

```python
df.drop('Id',axis=1,inplace=True)
```

```python
model_para={'svm':{'model':SVC(gamma='auto'),'params':{'C':[1,10,20],'kernel':['rbf','linear']}},
           'random_forest':{'model':RandomForestClassifier(),'params':{'n_estimators':[1,5,10]}},
           'logistic_reg':{"model":LogisticRegression(),'params':{'C':[1,5,10]}}}
```

```python
score=[]
models=['svm','random_forest','logistic_reg']
for model_name in models:
    mp=model_para[model_name]
    gds=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=True)
    gds.fit(iris.data,iris.target)
    score.append({'model':model_name,'best_score':gds.best_score_,'best_params':gds.best_params_})
```

```python
score
```

# digits example

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.datasets import load_digits
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
digits=load_digits()
```

```python
model_para={'svm':{'model':SVC(gamma='auto'),'params':{'C':[1,10,20],'kernel':['rbf','linear']}},
           'random_forest':{'model':RandomForestClassifier(),'params':{'n_estimators':[1,5,10]}},
           'logistic_reg':{"model":LogisticRegression(),'params':{'C':[1,5,10]}},
           'naive':{'model':MultinomialNB(),'params':{'alpha':[1.0]}},
           'deci_tree':{'model':DecisionTreeClassifier(),'params':{'criterion': ['gini','entropy']}}}
```

```python
score=[]
models=['svm','random_forest','logistic_reg','naive','deci_tree']
for model_name in models:
    mp2=model_para[model_name]
    gds=GridSearchCV(mp2['model'],mp2['params'],cv=5,return_train_score=True)
    gds.fit(digits.data,digits.target)
    score.append({'model':model_name,'best_score':gds.best_score_,'best_params':gds.best_params_})
```

```python
score
```
